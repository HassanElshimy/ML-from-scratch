- Create a vector [3, -2]
- Create a second vector [1, 4] and add it to the first
- Multiply your first vector by 2 and by -1
- Compute the magnitude of your first vector
- Create a 3D vector [1, 2, 3], add it to another 3D vector, and scale it

```python
import numpy as np
v1 = np.array([3,-2])
v2= np.array([1,4])
print(v1,v2)

addition_1 = v1 + v2
print(f"The addition of the above vectors is {addition_1}")

v1_scaled_1 = v1*2
v1_scaled_2 = v1*-1
print(v1_scaled_1,v1_scaled_2)

v1_magnitude = np.linalg.norm(v1)
print(f"The magnitude of the first vector is {v1_magnitude}")

v3 = np.array([1,2,3])
v4 = np.array([9,4,-1])
print(v3, v4)
addition_2 = v3 + v4
print(f"The addition of the above vectors is {addition_2}")

v3_scaled = v3 * -3
print(v3_scaled)
```
